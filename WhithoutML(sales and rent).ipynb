```python
#Works in sales and rent, just change the filename

import pandas as pd
# df = pd.read_csv("rent_prices.csv")
df = pd.read_csv("sale_prices.csv")

df = df.fillna(0)

raw_prices = df["Price on site"].tolist()
 
float_min_price = [float(price) for price in df["Min price"]]
float_max_price = [float(price) for price in df["Max price"]]
dictionary='1234567890'
min_prices = []
max_prices = []

for raw_price in raw_prices:
    parts = str(raw_price).split(' ') # split unparced price by space symbol and find sections of digits 
    prices=[]
    for part in parts:
        to_append = ''
        for s in part:
            if s in dictionary or s=='.':
                to_append+=s
        if to_append!='':         # check if parsed price is not empty and contains digits  
            for s in to_append:      #to avoid cases when there are only dots  
                if s in dictionary:
                    prices.append(to_append)
                    break
                    
    if prices: # if numbers were found
        min_prices.append(float(prices[0]))
        if len(prices)==2:
            max_prices.append(float(prices[1]))
        elif '+' in str(raw_price) or 'rom' in str(raw_price): #Check if there are From(from) word, or + symbol
            max_prices.append(0.0)                             #If there are - max price is unknown, so make it 0
        else:
            max_prices.append(float(prices[0]))
    else:                                                 #If we didn't found any sections of digits, price is unknown
        min_prices.append(0.0)
        max_prices.append(0.0)

        
for i in range(len(min_prices)):
    if(min_prices[i]!=float_min_price[i] or max_prices[i]!=float_max_price[i]):
        print(raw_prices[i])
        print(min_prices[i])
        print(max_prices[i])
        print(float_min_price[i])
        print(float_max_price[i])



     
                
            
```
